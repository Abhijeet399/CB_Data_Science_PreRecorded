```python
u = np.mean(dataset, axis = 0)
std = np.std(dataset, axis = 0)
```

```python
dataset = (dataset - u)/std
```

```python
ones = np.ones((dataset.shape[0], 1))
dataset = np.hstack((ones, dataset))
```

```python
split = 0.8

train_X = dataset[:int(split * dataset.shape[0]), :]
train_Y = label[:int(split * dataset.shape[0])]
test_X = dataset[int(split * dataset.shape[0]):, :]
test_Y = label[int(split * dataset.shape[0]):]
```

```python
print(test_X.shape, test_Y.shape, train_X.shape, train_Y.shape)
```

```python
def hypothesis(X,theta):
    return 1 / (1 + np.exp(-1 * np.dot(X, theta)))
```

```python
def loss_func(X, Y, weights):
    
    h_x = hypothesis(X, weights)
    loss = np.mean(Y*np.log(h_x) + (1 - Y)*np.log((1 - h_x)))
    
    return -1*loss
```

```python
def gradient(x, y, weight):
    
    y_pred = hypothesis(x, weight)
    grad = np.dot(x.T, (y - y_pred))
        
    return grad/x.shape[0]
```

```python
def gradient_descent(x, y, learning_rate = 0.03):
    
    weight = np.zeros((x.shape[1], ))
    
    cost = []
    
    epoch = 10000
    
    for i in range(epoch):
        grad = gradient(x, y, weight)
        
        loss = loss_func(x, y, weight)
        cost.append(loss)

        weight = weight + learning_rate * grad
        
    return cost, weight
```

```python
cost, weight = gradient_descent(train_X, train_Y)
```

```python
plt.plot(cost)
```

```python
def preict(X, weights):
    pred = hypothesis(X, weights)
    if pred >= 0.5:
        return 1
    else:
        return 0
```

```python
corr = 0

for i in range(test_X.shape[0]):
    pred = preict(test_X[i], weight)
    if pred == test_Y[i]:
        corr += 1
print(corr/test_X.shape[0])
```
